# Computer Vision - Image Plotting

This notebook shows the basics of plotting images using OpenCV and Matplotlib.



Steps:
1. Read image and convert to RGB
2. Plot image with rectangle
3. Plot cutted image
4. Plot grid



### Import Libraries

```python
# Import plotting libraries
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Import Image processing libraries
import cv2
```

### Read Image

```python
!ls ../img
```

```python
# Read image
file = "../img/image.jpg"
file = "../img/unsplash-mountain.jpg"
file = "../img/unsplash-sea.jpg"
# file = "img/color.png"
img = cv2.imread(file)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert to RGB for matplotlib

# CV2 uses BGR, matplotlib uses RGB
# CV2 reads image as numpy array (height, width, channels)

print("Image shape: ", img.shape)
```

### Plot Image

```python

rectangle_kwargs = dict(linewidth=3, edgecolor='lime', facecolor='none')

# First cut
dx, dy = 1350, 1080
x1, y1 = 2450, 600
x2, y2 = x1 + dx, y1 + dy
rect1 = Rectangle((x1, y1), dx, dy, **rectangle_kwargs)
img_cut1= img[y1:y2, x1:x2]


# Second cut
dx, dy = 200, 200
x1, y1 = 200, 200
x2, y2 = x1 + dx, y1 + dy
rect2 = Rectangle((x1, y1), dx, dy, **rectangle_kwargs)
img_cut2= img_cut1[y1:y2, x1:x2]


# Third cut
dx, dy = 30, 30
x1, y1 = 115, 72
x2, y2 = x1 + dx, y1 + dy
rect3 = Rectangle((x1, y1), dx, dy, **rectangle_kwargs)
img_cut3= img_cut2[y1:y2, x1:x2]


# Plot images (1 row, 4 columns)
fig, axs = plt.subplots(1, 4, figsize=(20, 30))


# Plot grid and title
for ax, title in zip(axs, ("Original image (x1, y1)=(2450, 600))", "Cut image", "Cut image 2", "Cut 3")):
    ax.grid(True, linestyle=':', linewidth=1, color='r')
    ax.set_title(title)


# Plot original image with rectangle
axs[0].imshow(img)
axs[0].add_patch(rect1)

# Add labet to x and y axis
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")



# Cut 1
axs[1].imshow(img_cut1)
axs[1].add_patch(rect2)

# Cut 2
axs[2].imshow(img_cut2)
axs[2].add_patch(rect3)

# Cut 3
axs[3].imshow(img_cut3)
# axs[3].add_patch(rect3)

plt.show()
```
